```python
import yfinance as yf
from pandas import MultiIndex, Int16Dtype
from datetime import datetime, timedelta
import datetime
from datetime import *
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARIMA
import numpy as np
import seaborn as sns
from tqdm import tqdm
import pandas as pd
from statsmodels.tools.sm_exceptions import ValueWarning, HessianInversionWarning, ConvergenceWarning
import warnings
import pandas_datareader as web 
```

```python
tickerSymbol1 = 'AVGO'
data1 = yf.Ticker(tickerSymbol1)
raw1 = data1.history(interval="1d",start = "2023-01-01", end = "2023-07-10")
stock1 = raw1["Close"].pct_change().dropna().to_numpy()
print(stock1.shape)
```

```python
tickerSymbol2 = 'AMZN'
data2 = yf.Ticker(tickerSymbol2)
raw2 = data2.history(interval="1d",start = "2023-01-01", end = "2023-07-10")
stock2 = raw2["Close"].pct_change().dropna().to_numpy()
print(stock2.shape)
```

```python
tickerSymbol3 = 'ULTA'
data3 = yf.Ticker(tickerSymbol3)
raw3 = data3.history(interval="1d",start = "2023-01-01", end = "2023-07-10")
stock3 = raw3["Close"].pct_change().dropna().to_numpy()
print(stock3.shape)
```

```python
corr = np.corrcoef((stock1, stock2, stock3))
print(corr)
plt.figure(figsize=(13, 8))
sns.heatmap(corr, annot = True,cmap= "RdYlGn")
plt.figure()
```
